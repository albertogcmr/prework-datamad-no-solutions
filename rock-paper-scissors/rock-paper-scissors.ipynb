```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
juego = ["stone","paper","scissors"]

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
maximo = 5

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
victoria = 3

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def jugada_aleatoria(juego):
    return random.choice(juego)

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def menu(juego):
    salir = False
    jugada_jugador = ""
    while not salir:
        comando = input("Introduzca el numero correspondiente a la opcion que desea:\n1.Stone\n2.Paper\n3.Scissors\n")
        if comando == "1":
            jugada_jugador = juego[0]
            salir = True
        elif comando == "2":
            jugada_jugador = juego[1]
            salir = True
        elif comando == "3":
            jugada_jugador = juego[2]
            salir = True
        else:
            print("Comando incorrecto\n")
    return jugada_jugador
# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def combate(jugada_jugador,jugada_maquina):
        resultado = 0
        if jugada_jugador == "stone":
            if jugada_maquina == "paper":
                resultado = 1
            elif jugada_maquina == "scissors":
                resultado = 2       
        elif jugada_jugador == "paper":
            if jugada_maquina == "stone":
                resultado = 2
            elif jugada_maquina == "scissors":
                resultado = 1
        elif jugada_jugador == "scissors":
            if jugada_maquina == "stone":
                resultado = 1
            elif jugada_maquina == "paper":
                resultado = 2
        return resultado
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def estado_juego(jugada_maquina,jugada_jugador,resultado,puntos_maquina,puntos_jugador):
    print("Maquina:",jugada_maquina,"Jugador:",jugada_jugador)
    if resultado == 0:
        print("Empate")
    elif resultado == 1:
        print("Maquina gana")
    else:
        print("Jugador gana")
    print("Puntos Jugador:",puntos_jugador,"Puntos maquina:",puntos_maquina,"\n")
    
# Create two variables that accumulate the wins of each participant
puntos_jugador = 0
puntos_maquina = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

partidas = 0

while puntos_jugador < 3 and puntos_maquina < 3 and partidas < maximo:
    jugada_maquina = jugada_aleatoria(juego)
    jugada_jugador = menu(juego)
    resultado = combate(jugada_jugador,jugada_maquina)
    
    if resultado == 1:
        puntos_maquina = puntos_maquina +1
        
    elif resultado == 2:
        puntos_jugador = puntos_jugador +1
        
    
    estado_juego(jugada_maquina,jugada_jugador,resultado,puntos_maquina,puntos_jugador)   
    partidas = partidas +1
# Print by console the winner of the game based on who has more accumulated wins
if puntos_jugador > puntos_maquina:
    print("Jugador gana la partida!")
elif puntos_jugador < puntos_maquina:
    print("Maquina gana la partida!")
else:
    print("Empate")
```
